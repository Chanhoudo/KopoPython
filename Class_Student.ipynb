```python
import random

grade_list = ['1학년', '2학년','3학년','4학년']
s_class = ['1반', '2반', '3반', '4반', '5반', '6반']
subject_list = ['국어','영어','수학']
student_c = 20

for i in range(len(grade_list)):
  for j in range(len(s_class)):
    for k in range(len(subject_list)):
      score_list = {}
      with open(f"{grade_list[i]}_{s_class[j]}_{subject_list[k]}", 'w', encoding = 'utf-8') as f:
        for m in range(1, student_c+1):
          score = random.randint(0, 100)
          f.write(str(score) + ' ')
```

```python
def print_menu():
  print('==================================')
  print('1. 학년 평균')
  print('2. 반 평균')
  print('3. 학생 등수')
  print('q. 종료')
  print('==================================')

```

```python

class Students:
  class_list = ['1반', '2반', '3반', '4반', '5반', '6반']
  student_num = 20
  sub_list = ['국어', '영어', '수학']


  def __init__(self, grade, class_n):
    # with open(f"{grade}_{n_class}_국어", 'r', encoding = 'utf-8') as f:
    #   kor_sub = f.readline().split(' ')
    # with open(f"{grade}_{n_class}_영어", 'r', encoding = 'utf-8') as f:
    #   eng_sub = f.readline().split(' ')
    # with open(f"{grade}_{n_class}_수학", 'r', encoding = 'utf-8') as f:
    #   mat_sub = f.readline().split(' ')
    self.grade = grade
    self.class_n = class_n

  def grade_avg(self):
    sum_list = [0,0,0]
    avg_dict = {}
    for i in range(len(self.class_list)):
      for j in range(len(self.sub_list)):
        self.sub_score = {'국어' : 0, '수학' : 0, '영어' : 0}
        with open(f"{self.grade}_{self.class_list[i]}_{self.sub_list[j]}", 'r', encoding = 'utf-8') as f:
          self.sub_score[self.sub_list[j]] = f.readline().split(' ')
          self.sub_score[self.sub_list[j]].pop()
        sum = 0
        for k in range(self.student_num):
          sum += int(self.sub_score[self.sub_list[j]][k])
        sum_list[j] += sum
    for i in range(len(sum_list)):
      avg_dict[self.sub_list[i]] = sum_list[i] / (self.student_num * len(self.class_list))
      print(f"{self.sub_list[i]} 과목의 평균 : {avg_dict[self.sub_list[i]]}")

  def class_avg(self):
    sum_dict = {}
    avg_dict = {}
    for j in range(len(self.sub_list)):
        self.sub_score = {'국어' : 0, '수학' : 0, '영어' : 0}
        with open(f"{self.grade}_{self.class_n}_{self.sub_list[j]}", 'r', encoding = 'utf-8') as f:
          self.sub_score[self.sub_list[j]] = f.readline().split(' ')
          self.sub_score[self.sub_list[j]].pop()
        sum = 0
        for k in range(self.student_num):
          sum += int(self.sub_score[self.sub_list[j]][k])
        sum_dict[self.sub_list[j]] = sum
    for i in range(len(sum_dict)):
      avg_dict[self.sub_list[i]] = sum_dict[self.sub_list[i]] / (self.student_num)
      print(f"{self.sub_list[i]} 과목의 평균 : {avg_dict[self.sub_list[i]]}")

  def student_rank(self):
    number = int(input('학생의 번호를 입력해주세요(1, 2, 3, 4 ... 20): '))
    rank_list = []
    for i in range(len(self.sub_list)):
      rank = 1
      with open(f"{self.grade}_{self.class_n}_{self.sub_list[i]}", "r", encoding = 'utf-8') as f:
        self.sub_score[self.sub_list[i]] = f.readline().split(' ')
        self.sub_score[self.sub_list[i]].pop()
      for j in range(self.student_num):
        if self.sub_score[self.sub_list[i]][j] > self.sub_score[self.sub_list[i]][number-1]:
          rank += 1
      print(f"{self.sub_list[i]} 과목의 등수 : {rank}")
      #rank_list.append(rank)


```

```python
grade_n = input('학년을 입력해주세요 :')
class_n = input('반을 입력해주세요 : ')

student_m1 = Students(grade_n, class_n)

while True:
  print_menu()
  menu = input('원하시는 항목을 선택해주세요: ')
  if menu == '1':
    student_m1.grade_avg()
  elif menu == '2':
    student_m1.class_avg()
  elif menu == '3':
    student_m1.student_rank()
  elif menu == 'q':
    break
```
