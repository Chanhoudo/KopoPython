```python
#1. 김소평

A = int(input("첫번째 숫자를 입력하세요"))
B = int(input("두번째 숫자를 입력하세요"))

C = A + B
print(A, "+", B, "=", C)

C = A - B
print(A, "-", B, "=", C)
```

```python
#2. 김영욱

N = int(input("N입력"))

SUM = 0
A = 1
while A <= N:
  B = A*A
  SUM += B
  print(SUM)
  A += 1
```

```python
#김영욱 // 결과값만 출력하는 버전

N = int(input("N입력"))

SUM = 0
A = 1
while A <= N:
  B = A*A
  SUM += B
  A += 1

print(SUM)
```

```python
#8. 이정현

len = int(input("거리를 입력(단위: km):"))

if len >= 40:
  age = int(input("나이를 입력: "))
  gender = input("성별을 입력: ")
  name = input("이름을 입력: ")
  eu = input("출퇴근 거리가 먼 이유(?)를 입력: ")
  print("거리는",len, "나이는", age, "성별은", gender, "이름은", name, "먼 이유는", eu)
else:
  age = int(input("나이를 입력: "))
  gender = input("성별을 입력: ")
  name = input("이름을 입력: ")
  print("거리는",len, "나이는", age, "성별은", gender, "이름은", name)
```

```python
#이정현 // age, gender, name은 동일하게 받기 때문에 미리 받고 40km가 넘으면 그때 eu를 추가로 인터뷰 함.

len = int(input("거리를 입력(단위: km):"))
age = int(input("나이를 입력: "))
gender = input("성별을 입력: ")
name = input("이름을 입력: ")
eu = 0

if len >= 40:
  eu = input("출퇴근 거리가 먼 이유(?)를 입력: ")
  print("거리는",len, "나이는", age, "성별은", gender, "이름은", name, "먼 이유는", eu)
else:
  print("거리는",len, "나이는", age, "성별은", gender, "이름은", name)
```

```python
#9. 정봉준 // 5번 반복 시키려면 [반복 조건을 i < 5]  또는 [i = 1부터 시작]

i = 0
while i < 5:
  r = int(input("반지름을 입력해주세요: "))
  o = r * r * 3.14
  print(f"{r} * {r} * 3.14 = {o}")
  i = i + 1
```

```python
#11. 추예진
answer = 1234
c = 0

while 1:
  p = int(input("비밀번호를 입력하세요(숫자만): "))
  if p == answer:
    print("잠금해체")
    break
  else:
    print("비밀번호를 틀리셨습니다. \n","5회 오류시 전원이 꺼집니다.")
    c += 1
    if(c >= 5):
      break
```

```python
#12. 황선균
weight = int(input("무게판별기"))

def weight_go(a):
  if (a)<=50 :
    print ("오른쪽으로")
  elif 50 < a <= 100 :
    print ("왼쪽으로")
  else:
    print("직진으로")

weight_go(weight)
```
