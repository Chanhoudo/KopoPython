```python
#인덱스 접근해서 출력하기 (월화수목금 => 1,2,3,4,0를 딕셔너리를 이용.)
#차량 뒷 번호와 요일을 잘 입력했는지 체크하기

car_stop = [[True,False,False,False,False],
            [False,True,False,False,False],
            [False,False,True,False,False],
            [False,False,False,True,False],
            [False,False,False,False,True]]
day_dic = {"월" : 1, "화" : 2, "수" : 3, "목" : 4, "금" : 0}
day_list = ["월", "화", "수", "목", "금"]


while True:
  car_end_num = int(input("차량 끝 번호 입력(0,9): "))
  if 0 <= car_end_num <= 9:
    break
  print("차량 끝 번호가 잘못 됐습니다.")

while True:
  day = input("오늘의 일을 입력(월,화,수,목,금): ")
  if day in day_list:
    break
  print("요일을 잘 못 입력하셨습니다. (월,화,수,목,금) 중에 선택해주세요")

car_end_num = car_end_num % 5

if car_stop[car_end_num][day_dic[day]]:
  print("대중교통을 이용해주세요~")
else:
  print("안전운전 하세요~")
```
