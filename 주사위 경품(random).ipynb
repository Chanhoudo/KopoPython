```python
import random

dice_dic = {1:"벤츠", 2:"호텔", 3:"100만원", 4:"꽝", 5:"유람선", 6:"발레니 30"}
price = int(input("지갑에 있는 현찰 입력(원 단위): "))

while True:
  answer = input("주사위를 던지시겠습니까(회 당 5000원)?(Y,N)")

  if answer == 'Y':
    if price < 5000:
      print("돈이 부족하시네요 ㅠ")
      break
    dice = random.randint(1,6)
    print(dice_dic[dice], "당첨!!")
    price -= 5000

  elif answer == 'N':
    print("다음에 또 만나요~")
    break

  else:
    print("(Y,N)만 입력해주세요..")

```
