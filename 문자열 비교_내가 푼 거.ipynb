```python
def diff(answer, str):
  for i in range(len(answer)):
    if(answer[i] == str[i]):
      continue
    else:
      return i
  return 0

answer = "장원영은 예쁘다, 결혼하고 싶다!!"
str = input("장원영은 예쁘다, 결혼하고 싶다!! 를 입력해주세요.")

res = diff(answer, str)
if res == 0:
  print("정답입니다.")
else:
  print("실패입니다. 틀린 부분은" , res + 1, "번째 글자 입니다.")
  print("틀린 글자는: ", str[res], "입니다.")
```
