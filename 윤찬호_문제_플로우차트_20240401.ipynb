```python
#17 윤홍림
A_list = []
B_list = []

A = int(input("A입력: "))
B = int(input("B입력: "))

max_yack = 0
min_gong = 0

for n in range(1, A+1):
  if A % n == 0:
    A_list.append(n)
for n in range(1, B+1):
  if B % n == 0:
    B_list.append(n)

for i in range(len(A_list)-1, -1, -1):
  for j in range(len(B_list)-1, -1, -1):
    if A_list[i] == B_list[j]:
      max_yack = A_list[i]
      break
  if max_yack != 0:
    break
min_gong = int(A * B / max_yack)

print("최대 공약수:", max_yack, "최소 공배수: ", min_gong)

```

```python
#18 최수아

item = input("아이템 입력: ")
price = int(input("가격 입력: "))
count = int(input("반복 횟수 입력: "))
total = 0

while count < 5:
  total += price * count
  count += 1

print("아이템: "+item,"가격:", price,"갯수:", count,"총합:", total)
```

```python
#18 최수아 // 반복문 변경, 더하는 수 변경

item = input("아이템 입력: ")
price = int(input("가격 입력: "))
count = int(input("반복 횟수 입력: "))
total = 0

for i in range(count):
  total += price

print("아이템: "+item,"가격:", price,"갯수:", count,"총합:", total)
```
