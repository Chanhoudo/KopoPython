```python
#임의의 갯수 숫자
#op 가 +, -, x, / 받아와서 결과값 반환

def operation(*nums, op):
  answer = 0;
  for num in nums:
    if op == '+':
      answer += num
    elif op == '-':
      if answer == 0:
        answer = num
        continue
      answer -= num
    elif op == 'x':
      if answer == 0:
        answer = 1
      answer *= num
    elif op == '/':
      if answer == 0:
        answer = num
        continue
      answer /= num
  return answer;


#i = int(input("몇 개의 개수의 숫자를 받으시겠습니까? "))
oper = input("기호를 골라주세요(+, -, x, /): ")

print("asnwer: ",operation(1,2,3,4,5,6, op=oper))
```
