```python
#1학년 4개 반 , 6학년까지 4반씩, 각 반에는 학생 20명
#국영수 시험, 시험 점수는 랜덤함수로 임의로 (0~100) 1학년에 국어점수 평균
#함수 : 학년별 과목의 평균점수, 특정반에 과목의 평균점수, 학생의 등수, 학년별로 각 과목의 최고 점수와 최저 점수

def menu():
  print("-------------------------------------------------")
  print("--------------학생 평균보는 메뉴-----------------")
  print("-------------------------------------------------")
  print("1. 학년별 평균 확인(국, 영, 수)")
  print("2. 특정반 평균 확인(국, 영, 수)")
  print("3. 학생의 등수 확인")
  print("4. 학년별로 각 과목의 최고 점수와 최저 점수 확인")
  print("q. 종료")
  select = (input("원하시는 항목을 적어주세요: "))

  return select
```

```python
def grade_avg(students): #1번 학년별 평균
  grade = int(input("원하시는 학년을 입력: "))
  sum_list = [0,0,0]
  stack = 0
  for i in range(len(students[1])):
    for j in range(len(students[1][1])):
      for k in range(3):
        sum_list[k] += students[grade][i+1][j+1][k]
      stack += 1
  print(f"{grade}학년 학생들의 평균\n국어 평균: {int(sum_list[0]/stack)}\n영어 평균: {int(sum_list[2]/stack)}\n수학 평균: {int(sum_list[1]/stack)}\n")
```

```python
def class_avg(students): #2번 특정반 평균
  grade = int(input("원하는 학년을 입력: "))
  class1 = int(input("원하는 반을 입력: "))
  sum_list = [0,0,0]
  stack = 0
  for i in range(len(students[grade][class1])):
    sum_list[0] += students[grade][class1][i+1][0]
    sum_list[1] += students[grade][class1][i+1][1]
    sum_list[2] += students[grade][class1][i+1][2]
    stack += 1

  print(f"{grade}학년 {class1}반의 학생들의 평균\n국어 평균: {int(sum_list[0]/stack)}\n영어 평균: {int(sum_list[2]/stack)}\n수학 평균: {int(sum_list[1]/stack)}\n")
```

```python
def student_rank(studnets): #3번 학생별 등수 확인
  grade = int(input("등수를 보기를 원하는 학생의 학년을 입력: "))
  class1 = int(input("등수를 보기 원하는 학생의 반을 입력: "))
  number = int(input("등수를 보기 원하는 학생의 학생의 번호를 입력: "))

  rank_list = [1,1,1]

  for i in range(len(students[grade][class1])):
    if students[grade][class1][number][0] < students[grade][class1][i+1][0]:
      rank_list[0] += 1
    if students[grade][class1][number][1] < students[grade][class1][i+1][1]:
      rank_list[1] += 1
    if students[grade][class1][number][2] < students[grade][class1][i+1][2]:
      rank_list[2] += 1

  print(f"{number}번의 학생의 등수\n 국어: {rank_list[0]}등\n 수학: {rank_list[1]}등\n 영어: {rank_list[2]}등  입니다.")
```

```python
def m_m_score(students):
  grade = int(input("등수를 보기를 원하는 학생의 학년을 입력: "))

  max_list = [students[grade][1][1][0], students[grade][1][1][1], students[grade][1][1][2]]
  min_list = [students[grade][1][1][0], students[grade][1][1][1], students[grade][1][1][2]]

  for i in range(len(students[grade])):
    for j in range(1, len(students[grade][1])):
      if max_list[0] < students[grade][i+1][j+1][0]:
        max_list[0] = students[grade][i+1][j+1][0]
      if max_list[1] < students[grade][i+1][j+1][1]:
        max_list[1] = students[grade][i+1][j+1][1]
      if max_list[2] < students[grade][i+1][j+1][2]:
        max_list[2] = students[grade][i+1][j+1][2]

      if min_list[0] > students[grade][i+1][j+1][0]:
        min_list[0] = students[grade][i+1][j+1][0]
      if min_list[1] > students[grade][i+1][j+1][1]:
        min_list[1] = students[grade][i+1][j+1][1]
      if min_list[2] > students[grade][i+1][j+1][2]:
        min_list[2] = students[grade][i+1][j+1][2]
  print(f"=====최고점수 리스트=====\n국어 최고 점수: {max_list[0]}\n영어 최고 점수: {max_list[1]}\n수학 최고 점수: {max_list[2]}\n\n국어 최저 점수: {min_list[0]}\n영어 최저 점수: {min_list[1]}\n수학 최저 점수: {min_list[2]}")
```
