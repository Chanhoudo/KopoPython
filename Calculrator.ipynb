```python
def menu_list():
  print('=========================================')
  print('================계산기===================')
  print('=========================================')
  print('1. 더하기')
  print('2. 빼기')
  print('3. 곱하기')
  print('4. 나누기')
  print('5. 숫자 변경')
  print('q. 종료')
```

```python
class Calculrator:

  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2

  def change_num(self):
    self.num1 = int(input('첫번째 정수를 입력해주세요: '))
    self.num2 = int(input('두번째 정수를 입력해주세요: '))

  def plus(self):
    return self.num1 + self.num2
  def mul(self):
    print('절대값으로 출력됩니다.')
    return abs(self.num1 - self.num2)
  def gob(self):
    return self.num1 * self.num2
  def div(self):
    return self.num1 / self.num2


n1 = int(input('첫번째 정수를 입력해주세요: '))
n2 = int(input('두번째 정수를 입력해주세요: '))
Cal = Calculrator(n1,n2)

while True:
  menu_list()
  menu = input('선택: ')

  if menu == '1':
    print(Cal.plus())
  elif menu == '2':
    print(Cal.mul())
  elif menu == '3':
    print(Cal.gob())
  elif menu == '4':
    print(Cal.div())
  elif menu == '5':
    Cal.change_num()
  elif menu == 'q':
    break
  else:
    print("잘못 입력하셨습니다.")
```
