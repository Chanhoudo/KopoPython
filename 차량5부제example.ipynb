```python
#인덱스 접근해서 출력하기 (월화수목금 => 1,2,3,4,0를 딕셔너리를 이용.)
#차량 뒷 번호와 요일을 잘 입력했는지 체크하기
#dic in list

car_stop = {"금" : [True, False, False, False, False],
            "월" : [False, True, False, False, False],
            "화" : [False, False, True, False, False],
            "수" : [False, False, False, True, False],
            "목" : [False,False,False, False, True]}

while True:
  car_end_num = int(input("차량 끝 번호 입력(0,9): "))
  if 0 <= car_end_num <= 9:
    break
  print("차량 끝 번호가 잘못 됐습니다.")

while True:
  day = input("오늘의 일을 입력(월,화,수,목,금): ")
  if day in car_stop.keys():
    break
  print("요일을 잘 못 입력하셨습니다. (월,화,수,목,금) 중에 선택해주세요")

car_end_num = car_end_num % 5

if car_stop[day][car_end_num]:
  print("대중교통을 이용해주세요~")
else:
  print("안전운전 하세요~")
```

```python
#인덱스 접근해서 출력하기 (월화수목금 => 1,2,3,4,0를 딕셔너리를 이용.)
#차량 뒷 번호와 요일을 잘 입력했는지 체크하기
#list in dic

car_stop = [{"금" : 1, "화" : 0, "수" : 0, "목" : 0, "월" : 0},
            {"금" : 0, "화" : 1, "수" : 0, "목" : 0, "월" : 0},
            {"금" : 0, "화" : 0, "수" : 1, "목" : 0, "월" : 0},
            {"금" : 0, "화" : 0, "수" : 0, "목" : 1, "월" :0},
            {"금" : 0, "화" : 0, "수" : 0, "목" : 0, "월" : 1}]


while True:
  car_end_num = int(input("차량 끝 번호 입력(0,9): "))
  if 0 <= car_end_num <= 9:
    break
  print("차량 끝 번호가 잘못 됐습니다.")

while True:
  day = input("오늘의 일을 입력(월,화,수,목,금): ")
  if day in car_stop[0].keys():
    break
  print("요일을 잘 못 입력하셨습니다. (월,화,수,목,금) 중에 선택해주세요")

car_end_num = car_end_num % 5

if car_stop[car_end_num][day]:
  print("대중교통을 이용해주세요~")
else:
  print("안전운전 하세요~")
```
