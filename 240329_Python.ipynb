```python
# names라는 리스트 ["장원영, 장혁", "장보고", "김지원", " 김수현"]

names = ["장원영", "장혁", "장보고", "김지원", "김수현"]
jang = []
kim = []
name_len_two = []

jang = list(filter(lambda x : x[0] == "장", names))
kim = list(filter(lambda x : x[0] == "김", names))
name_len_two = list(filter(lambda x : len(x) == 2, names))

print("성이 장씨인 사람: ", jang)
print("성이 김씨인 사람: ", kim)
print("이름이 두 글자인 사람: ", name_len_two)
```
