```python
#처음 푼 방식
with open("test.txt", 'r', encoding = 'utf-8') as f:
  lines = f.readlines()

temp = ''
change_index = []

for line_i in range(len(lines)):
  for c_i in range(len(lines[line_i])):
    temp += lines[line_i][c_i]
    if len(temp) > 2:
      temp = temp[1] + temp[2]
    if temp == '영사':
      a_i = [line_i, c_i-1,'a']
      change_index.append(a_i)
    elif temp == '스드':
      a_i = [line_i, c_i-1,'b']
      change_index.append(a_i)

answer = ''
check_len = False
for line_num in range(len(lines)):
  num_temp = 0
  str_temp = ''
  for i in change_index:
    if i[0] != line_num:
      continue
    str_temp += lines[line_num][num_temp:i[1]]
    num_temp = i[1] + 2
    if i[2] == 'a':
      str_temp += 'a'
    else:
      str_temp += 'b'

  if len(str_temp) < len(lines[line_num]):
    if not len(str_temp) == 0:
      str_temp += lines[line_num][len(str_temp)+1:]
    else:
      str_temp += lines[line_num][len(str_temp):]
  answer += str_temp



with open("test.txt", 'w', encoding = 'utf-8') as f:
  f.write(answer)

```

```python
#파이썬 함수 이용해서 쉽게 풀기
with open("test.txt", 'w', encoding = 'utf-8') as f:
  f.write("장원영사기린스드메론\n")
  f.write("스드메의문단속\n")
  f.write("안녕하세요\n")

with open("test.txt", 'r', encoding = 'utf-8') as f:
  lines = f.readlines()

answer = ''
for line in lines:
  line = line.replace('영사', 'a')
  line = line.replace('스드', 'b')
  answer += line

with open("test.txt", 'w', encoding = 'utf-8') as f:
  f.write(answer)

with open("test.txt", 'r', encoding = 'utf-8') as f:
  lines = f.readlines()

print(lines)
```
