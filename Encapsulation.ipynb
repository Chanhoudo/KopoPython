```python
class Computer:
  max = 400
  def __init__(self):
    self.__maxprice = 900

  def sell(self):
    print(f"Selling Price: {self.__maxprice}")
    print(f"max : {self.max}")

  def setMaxPrice(self, price):
    self.__maxprice = price

c = Computer()

c.sell()

c.max = 600
c.__maxprice = 1000
c.sell()

c.setMaxPrice(1000)
c.sell()
```
